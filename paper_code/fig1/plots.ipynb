```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
```

```python
DATA_DIR = "/Users/spare/Desktop/BIODS215/project/wearables/Basis_Watch_Data_Processed/"
template = os.path.join(DATA_DIR, "Basis_{}.csv")
```

## Plot average weekly activity and average daily activity for first four subjects

```python
dfs = []

fig, a = plt.subplots(4, 1, figsize=(12, 6))
for i, x in enumerate(['001', '002', '003', '004']):
    if (len(dfs)) != 4:
        dfs.append(pd.read_csv(template.format(x), index_col=0))
    df = dfs[i]
    week_df = df.groupby(["weekday", "hour"]).accel_magnitude.mean().reset_index()
    a[i].plot(range(len(week_df)), week_df.accel_magnitude)
    a[i].set_xticks([])
    a[i].set_yticks([])

#plt.savefig("/Users/spare/Desktop/BIODS215/project/plots/weekly_examples.png", dpi=300)
plt.show()
```
